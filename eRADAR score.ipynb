```python
#eRADAR score 
def calculate_eradar_score(row):
    intercept = -11.83  
    coefficients = {
        'age': 0.11,
        'female': -0.10,
        'congestive_heart_failure': 0.28,
        'cerebrovascular_disease': 0.18,
        'diabetes_complex': 0.34,
        'diabetes_any': 0.06,
        'bmi_under_18.5': 0.29,
        'bmi_over_30': -0.11,
        'ed_visits': 0.34,
        'antidepressant_fills': 0.58,
        'chronic_pulmonary_disease': -0.11,
        'dx_hypothyoidism': 0.05,
        'renal_failure': -0.15,
        'lymphoma': 0.01,
        'solid_tumor': -0.17,
        'rheumatoid_arthritis': -0.05,
        'weight_loss': -0.34,
        'fluid_electrolyte_disorders': -0.07,
        'blood_loss_anemia': -0.25,
        'bipolar_disorder_psychoses': 0.46,
        'depression': 0.03,
        'traumatic_brain_injury': 0.20,
        'tobacco_use': -0.15,
        'atrial_fibrillation': 0.03,
        'gait_abnormality': 0.24,
        'high_bmi': -0.09
    }

    logit = intercept
    for feature, coef in coefficients.items():
        logit += coef * row[feature]

    return 1 / (1 + math.exp(-logit))

```
